```python
# Calculating the runs made in each inning of every match
total_score_df = delivery.groupby(['match_id', 'inning']).sum()['total_runs'].reset_index()
total_score_df
```

```python
# Calculating the runs made in the first innings of every match
total_score_df = total_score_df[total_score_df.inning == 1]
total_score_df
```

```python
# Merging match table and total_score_df table 
match_df = match.merge(total_score_df[['match_id', 'total_runs']], left_on = 'id', right_on = 'match_id')
match_df.head()
```

```python
# Finding out the teams in the competition
match_df['team1'].unique()
```

```python
# Teams in the latest edition of the Indian Premier League
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'
]
```

#### <i> The teams Delhi Daredevils and Deccan chargers have been relaunched in the league with the names Delhi Capitals and Sunrisers Hyderabad respectively, so we need to make that change to the dataset. 

#### <i> Also to only analyse the performance of the currently partitipating teams, we need to filter out the teams which are not a part of the competition anymore, namely Kochi Tuskers Kerala, Gujarat Lions, Pune Warriors and Rising Pune Supergiants.

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils', 'Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils', 'Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df.head()
```

```python
# Checking for current teams only
match_df['team1'].unique()
```

```python
match_df['team2'].unique()
```

```python
match_df.shape
```

```python
match_df = match_df[match_df['dl_applied'] == 0]
```

```python
match_df = match_df[['match_id', 'city', 'winner', 'total_runs']]
```

```python
match_df.head()
```

```python
delivery.head()
```

```python
delivery_df = match_df.merge(delivery, on = 'match_id')
```

```python
delivery_df.head()
```

```python
delivery_df = delivery_df[delivery_df['inning'] == 2]
```

```python
delivery_df.head()
```

```python
delivery_df['current_score'] = delivery_df.groupby('match_id').cumsum()['total_runs_y']
```

```python
delivery_df.head()
```

```python
delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score']
```

```python
delivery_df.head()
```

```python
delivery_df['balls_left'] = 126 - (delivery_df['over'] * 6 + delivery_df['ball'])
```

```python
delivery_df.head()
```

```python
delivery_df.columns
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
wickets = delivery_df.groupby('match_id').cumsum()['player_dismissed'].values
delivery_df['wickets'] = 10 - wickets
delivery_df.head()
```

```python
# Current run rate(crr) = runs/overs
delivery_df['crr'] = (delivery_df['current_score'] * 6) / (120 - delivery_df['balls_left'])
```

```python
# Required run rate(rrr) = runs_left / overs_left
delivery_df['rrr'] = (delivery_df['runs_left'] * 6) / delivery_df['balls_left']
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result, axis = 1)
```

```python
delivery_df.head()
```

```python
final_df = delivery_df[['match_id','batting_team','bowling_team','city','runs_left','balls_left','wickets',
                        'total_runs_x','crr','rrr', 'result']]
```

```python
final_df.head()
```

```python
final_df.shape
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.sample()
```

```python
final_df.dropna(inplace = True)
```

```python
final_df = final_df[final_df['balls_left'] != 0]
```

```python
final_df.head()
```

```python
final_df.shape
```

```python
final_df.head()
```

```python
# Splitting the data into Training and Test data
X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse_output = False, drop = 'first'),['batting_team','bowling_team','city'])],
    remainder = 'passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps = [
    ('step1',trf),
    ('step2',LogisticRegression(solver = 'liblinear'))
])
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
